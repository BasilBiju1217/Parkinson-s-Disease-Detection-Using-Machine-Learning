# PAKINSONS DISEASE
## Context
Parkinson’s Disease (PD) is a degenerative neurological disorder marked by decreased dopamine levels in the brain. It manifests itself through a deterioration of movement, including the presence of tremors and stiffness. There is commonly a marked effect on speech, including dysarthria (difficulty articulating sounds), hypophonia (lowered volume), and monotone (reduced pitch range). Additionally, cognitive impairments and changes in mood can occur, and risk of
dementia is increased.

Traditional diagnosis of Parkinson’s Disease involves a clinician taking a neurological history of the patient and observing motor skills in various situations. Since there is no definitive laboratory test to diagnose PD, diagnosis is often difficult, particularly in the early stages when motor effects are not yet severe. Monitoring progression of the disease over time requires repeated clinic visits by the patient. An effective screening process, particularly one that doesn’t require a clinic visit, would be beneficial. Since PD patients exhibit characteristic vocal features, voice recordings are a useful and non-invasive tool for diagnosis. If machine learning algorithms could be applied to a voice recording dataset to accurately diagnosis PD, this would be an effective screening step prior to an appointment with a clinician.

Here’s a clear explanation of each column in the matrix you provided, which appears to be from a **Parkinson’s disease dataset based on voice measurements** (commonly used in machine learning and biomedical studies):

---

### **1. name**

* **Description**: Unique identifier for each subject, typically encoded as an ASCII string including the subject's name and a recording number.

---

### **Fundamental Frequency (Pitch) Measures**

These relate to how high or low a voice sounds.

### **2. MDVP\:Fo(Hz)**

* **Meaning**: Average fundamental frequency of the voice in Hertz (Hz).
* **Use**: Indicates average pitch of the voice.

### **3. MDVP\:Fhi(Hz)**

* **Meaning**: Maximum fundamental frequency (Hz).
* **Use**: Highest pitch observed.

### **4. MDVP\:Flo(Hz)**

* **Meaning**: Minimum fundamental frequency (Hz).
* **Use**: Lowest pitch observed.

---

### **Frequency Variation Measures (Jitter)**

These show *short-term* variations in pitch.

### **5. MDVP\:Jitter(%)**

* **Meaning**: Percentage of variation in pitch.
* **Use**: Higher values can indicate voice disorders.

### **6. MDVP\:Jitter(Abs)**

* **Meaning**: Absolute jitter in seconds.
* **Use**: Another pitch variation measure (not normalized).

### **7. MDVP\:RAP**

* **Meaning**: Relative Average Perturbation.
* **Use**: Measures average pitch perturbation over 3 cycles.

### **8. MDVP\:PPQ**

* **Meaning**: Five-point Period Perturbation Quotient.
* **Use**: Measures pitch variations over 5 cycles.

### **9. Jitter\:DDP**

* **Meaning**: Average absolute difference of differences between consecutive pitch periods (DDP = 3 × RAP).
* **Use**: A cumulative jitter measure.

---

### **Amplitude Variation Measures (Shimmer)**

These show *short-term* variations in loudness.

### **10. MDVP\:Shimmer**

* **Meaning**: Percentage variation in amplitude.
* **Use**: Indicates instability in voice loudness.

### **11. MDVP\:Shimmer(dB)**

* **Meaning**: Shimmer measured in decibels.
* **Use**: Loudness variation in decibels.

### **12. Shimmer\:APQ3**

* **Meaning**: Amplitude Perturbation Quotient over 3 cycles.
* **Use**: Like RAP but for amplitude.

### **13. Shimmer\:APQ5**

* **Meaning**: Amplitude Perturbation Quotient over 5 cycles.
* **Use**: Shimmer measured over longer cycles.

### **14. MDVP\:APQ**

* **Meaning**: Average perturbation quotient.
* **Use**: A general shimmer measure.

### **15. Shimmer\:DDA**

* **Meaning**: Average absolute difference between consecutive differences in amplitude (DDA = 3 × APQ3).
* **Use**: Cumulative shimmer metric.

---

### **Noise-to-Harmonics Ratio**

These relate to the clarity or breathiness of voice.

### **16. NHR (Noise-to-Harmonics Ratio)**

* **Meaning**: Ratio of noise to tonal components.
* **Use**: Higher in noisy, breathy voices, often in Parkinson's.

### **17. HNR (Harmonics-to-Noise Ratio)**

* **Meaning**: Ratio of tonal to noise components.
* **Use**: Higher means clearer voice.

---

### **Status Label**

### **18. status**

* **Meaning**: Diagnosis label.

  * **1** = Parkinson's Disease
  * **0** = Healthy
* **Use**: Target variable for classification.

---

### **Nonlinear Dynamical Measures**

Measure complexity or irregularity in vocal signal.

### **19. RPDE (Recurrence Period Density Entropy)**

* **Meaning**: Entropy of recurring patterns.
* **Use**: Higher in more complex, irregular voices.

### **20. D2 (Correlation Dimension)**

* **Meaning**: Measures signal complexity.
* **Use**: Higher values show increased irregularity, possibly indicating pathology.

### **21. DFA (Detrended Fluctuation Analysis)**

* **Meaning**: Signal fractal scaling exponent.
* **Use**: Indicates fractal correlation in the signal; used in characterizing long-term signal correlations.

---

### **Nonlinear Frequency Variation Measures**

Advanced measures of pitch variability.

### **22. spread1**

* **Meaning**: First nonlinear spread measure.
* **Use**: Relates to pitch variation asymmetry.

### **23. spread2**

* **Meaning**: Second nonlinear spread measure.
* **Use**: Captures different aspects of frequency variation.

### **24. PPE (Pitch Period Entropy)**

* **Meaning**: Entropy of the pitch period distribution.
* **Use**: Quantifies pitch unpredictability.




ASCII String Meaning:
ASCII (American Standard Code for Information Interchange) is a character encoding standard.

An ASCII string is simply a text string made up of characters that belong to the ASCII character set (e.g., letters, numbers, symbols like _, -, etc.).

In your dataset:
The name column contains an ASCII string such as:
"phon_R01_S01_1"
This could mean:

phon – voice sample type (phonation)

R01 – recording session

S01 – subject number

1 – trial number or first recording

So, ASSI string seems to be a typo, and what you’re referring to is an ASCII string used for naming or identifying records.
